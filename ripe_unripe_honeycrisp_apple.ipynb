```python
import numpy as np
import pandas as pd
import cv2 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
from datetime import date
import datetime
from datetime import datetime as datetime1
```

```python
frame = cv2.imread("/content/drive/MyDrive/apples/Honeycrisp/h16.jpg", cv2.IMREAD_COLOR)
cv2_imshow(cv2.resize(frame,(250,250)))
hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
```

```python
#Ranging and masking of colors
low_red = np.array([145, 155, 84])
high_red = np.array([179, 255, 255])
red_mask = cv2.inRange(hsv_frame, low_red, high_red)
red = cv2.bitwise_and(frame, frame, mask=red_mask)
low_green = np.array([25, 52, 72])
high_green = np.array([102, 255, 255])
green_mask = cv2.inRange(hsv_frame, low_green, high_green)
green = cv2.bitwise_and(frame, frame, mask=green_mask)
low_yellow = np.array([18, 30, 0])
high_yellow = np.array([32, 255, 255])
yellow_mask = cv2.inRange(hsv_frame, low_yellow, high_yellow)
yellow = cv2.bitwise_and(frame, frame, mask=yellow_mask)
```

```python
plt.imshow(red)
plt.imshow(red_mask)
plt.imshow(green_mask)
plt.imshow(yellow_mask)
a=int(cv2.countNonZero(red_mask))
b=int(cv2.countNonZero(green_mask))
c=int(cv2.countNonZero(yellow_mask))
```

```python
#prediction for just ripe or unripe apples
arr = []
if b < a+c:
  arr.append("Ripe honeycrisp apple")
  print("Ripe honeycrisp apple")
elif b > a+c:
  arr.append("Raw or unripe honeycrisp apple")
  print("Raw or unripe honeycrisp apple")
```
